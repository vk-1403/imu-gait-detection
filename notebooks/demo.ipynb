# Peak-anchored detector demo

Loads one Pacini Panebianco trial, runs the recommended detector, and plots the gyroscope trace with detected events.

```python
import sys, os
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from src import detect_events, Params
```

```python
DATA = Path('..') / 'data' / 'pacini_real'
TRIAL = 1
FS = 200.0

rsh = pd.read_csv(DATA / f'pacini_T{TRIAL:02d}_rshank.csv')
out = detect_events(rsh['gyro_z_rad_s'].values, fs=FS, params=Params(fs=FS))
print('HS:', len(out['hs_idx']), 'TO:', len(out['to_idx']))
```

```python
t = np.arange(len(out['gyro_filt'])) / FS
fig, ax = plt.subplots(figsize=(12, 3))
ax.plot(t, out['gyro_filt'], lw=0.8, color='#1F3B5B')
ax.scatter(t[out['hs_idx']], out['gyro_filt'][out['hs_idx']], color='#D55E00', s=18, label='HS')
ax.scatter(t[out['to_idx']], out['gyro_filt'][out['to_idx']], color='#0072B2', s=18, label='TO')
ax.set_xlabel('Time (s)')
ax.set_ylabel('omega_z (rad/s)')
ax.legend(loc='upper right', frameon=False)
ax.set_xlim(10, 25)
plt.tight_layout()
plt.show()
```
